```python
conditions = [
    (df['Revenue'] >= 2000),
    (df['Revenue'] >= 1500) & (df['Revenue'] < 2000),
    (df['Revenue'] < 1500)
]

choices = ['Excellent', 'Good', 'Needs Improvement']

df['grade'] = np.select(conditions, choices, default='No Grade')

df
```

```python
def get_grade(Revenue):
    if Revenue > 2000:
        return 'Excellent'
    elif Revenue >= 1500 and Revenue < 2000:
        return 'Better'
    elif Revenue >= 1000 and Revenue < 1500:
        return 'Good'
    else:
        return 'Needs Improvement'

df['grade_new'] = df['Revenue'].apply(get_grade)

df
```
