```python
import pandas as pd

# Define the file paths
input_file_path = r'F:\VI USER\My Work\Python\post paid bill.csv'
output_file_path = r'F:\VI USER\My Work\Python\output.csv'

# Read the input CSV file
df = pd.read_csv(input_file_path)

# Find unique values in the 'CUSTOMER_CELL_NUMBER' column
unique_numbers = df['CUSTOMER_CELL_NUMBER'].value_counts()
unique_numbers = unique_numbers[unique_numbers == 1].index.tolist()

# Filter out rows with unique 'CUSTOMER_CELL_NUMBER' values
filtered_df = df[~df['CUSTOMER_CELL_NUMBER'].isin(unique_numbers)]

# Apply additional filters
filter1 = filtered_df['MEDIA_CHANNEL'] == 'IBCC'
filter2 = filtered_df['SUB_MEDIA'].isin(['Postpaid CC'])
filter3 = filtered_df['SR_CATEGORY'] == 'Enquiry'
filter4 = filtered_df['SR_TYPE'] == 'Billing and Payment'
filter5 = filtered_df['SR_SUB_TYPE'] == 'Information'
filter6 = filtered_df['SR_SUB_SUB_TYPE'] == 'Explanation Of Bill'
filter7 = filtered_df['DFF_VAL2'] == 'Late fees'
filter8 = filtered_df['CIRCLE'].isin(['Madhya Pradesh', 'Rajasthan', 'UP East', 'UP West', 'Haryana', 'Delhi', 'Mumbai'])

# Apply the filters
filtered_df = filtered_df[filter1 & filter2 & filter3 & filter4 & filter5 & filter6 & filter7 & filter8]

# Remove duplicates in 'CUSTOMER_CELL_NUMBER' column
filtered_df = filtered_df.drop_duplicates(subset=['CUSTOMER_CELL_NUMBER'])

# Write the filtered DataFrame to a new CSV file
filtered_df.to_csv(output_file_path, index=False)
```
