```python
#Duplicates in DataFrames
```

```python
players = pd.DataFrame({
    'Name': [
        'Kobe Bryant',
        'LeBron James',
        'Kobe Bryant',
        'Carmelo Anthony',
        'Kobe Bryant',
    ],
    'Pos': [
        'SG',
        'SF',
        'SG',
        'SF',
        'SF'
    ]
})
```

```python
players
```

```python
players.duplicated()
```

```python
players.duplicated(subset=['Name'])
```

```python
players.duplicated(subset=['Name'], keep='last')
```

```python
players.drop_duplicates()
```

```python
players.drop_duplicates(subset=['Name'])
```

```python
df = pd.DataFrame({
    'Data': [
        '1987_M_US _1',
        '1990?_M_UK_1',
        '1992_F_US_2',
        '1970?_M_   IT_1',
        '1985_F_I  T_2'
]})
```

```python
df
```

```python

df['Data'].str.split('_')
```

```python
df['Data'].str.split('_', expand=True)
```

```python
df = df['Data'].str.split('_', expand=True)
```

```python
df.columns = ['Year', 'Sex', 'Country', 'No. Of Children']
```

```python
df
```

```python
df['Year'].str.contains('/?')
```

```python
df['Country'].str.contains('U')
```

```python
df['Country'].str.strip()
```

```python
df['Country'].str.replace(' ', '')
```

```python
df['Year'].str.replace('?', '')
```

```python
df["Year"]
```

```python
df['Year'].str.replace(r'(?P\d{4})\?', lambda m: m.group('year'))
```
