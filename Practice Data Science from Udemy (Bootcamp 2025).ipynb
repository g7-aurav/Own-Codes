The Variable Rand 1,2,3 not only worsen the explanatory power but is also insignificant

Assumptions

Here we divide them into five assumptions.You should know all of them and consider them before you perform regression analysis.

1.) The first one is linearity.
It is called a linear regression.
As we said earlier, there are other types
of regressions with more sophisticated models.
The linear regression is the simplest one
and assumes linearity.
Each independent variable is multiplied
by a coefficient and summed up to predict the value.

2.) The second one is endogeneity of regressors.
Mathematically, this is expressed as the covariance
of the error, and the Xs is 0 for any error or X.

3.) The third assumption is normality
and homoscedasticity of the error term.
Normality means the error term is normally distributed.
The expected value of the error is zero
as we expect to have no errors on average.
Homoscedasticity in plain English means constant variance.
We will discuss it thoroughly later.

4.) The fourth assumption is no autocorrelation.
Mathematically, the covariance
of any two error terms is zero.
That's the assumption that would usually stop you
from using a linear regression in your analysis.

5.) And the last one is no multicollinearity.
Multicollinearity is observed when two or more variables
have a high correlation between each other.
Okay, these are the main regression assumptions.
They are crucial for regression analysis.

As the purpose of this section
is to help you understand regressions,
you should not skip any lectures about assumptions.
At your workplace, the biggest mistake you can make
is to perform a regression that violates
one of these assumptions.
