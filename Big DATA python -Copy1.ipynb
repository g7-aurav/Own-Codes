```python
# Get data
CBDATA = combined_df['CIRCLE_NAME'].value_counts()

# Sort data alphabetically
CBDATA = CBDATA.sort_index()

# Calculate percentage contribution
percentages = CBDATA / CBDATA.sum() * 100

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(35,10))

# Create barplot with Seaborn
sns.barplot(y=CBDATA, x=CBDATA.index, ax=ax, palette='flare')

# Rotate x-axis label to 90 degrees
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

# Add data labels to bar plot with Matplotlib
for i, v in enumerate(CBDATA):
    label = str(round(v/1000, 2)) + 'K'
    ax.text(i, v + 0.5, label, color='black', fontweight='bold', fontsize=20, ha='center',bbox=dict(facecolor='white', alpha=0.9))

# Create a twin axis
ax2 = ax.twinx()

# Create line plot with Seaborn
sns.lineplot(x=percentages.index, y=percentages, ax=ax2, color='black', linewidth=3)

# Add data labels to the line plot with Matplotlib
for i, v in enumerate(percentages):
    label = str(round(v, 2)) + '%'
    ax2.annotate(label, (i, v), xytext=(5, 5), textcoords='offset points', color='white', fontsize=20, ha='center', va='center', fontweight='bold', bbox=dict(facecolor='blue', alpha=0.5))

# Customize the plot
ax.set_title("Highest Ratings", size=20, pad=20)
ax.set_xlabel("Circle", size=25)
ax.set_ylabel("Count", size=25)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=25, fontweight='light')
ax.set_yticklabels(ax.get_yticklabels(), fontsize=20, fontweight='light')
ax2.set_ylabel("Percentage Contribution", size=25, color='black')
ax2.tick_params(axis='y', labelsize=20, labelcolor='black')
ax2.set_ylim(-35,100)

# Show the plot
plt.show()
```

```python
import plotly.graph_objects as go

# Get data
chains = combined_df['CIRCLE_NAME'].value_counts()
data = pd.DataFrame({'Circle': chains.index, 'Count': chains.values})

# Calculate contribution percentage
total_count = data['Count'].sum()
data['Contribution'] = (data['Count'] / total_count) * 100

# Create bar plot using Plotly
fig = go.Figure()

# Add bar trace with data labels
fig.add_trace(go.Bar(
    x=data['Circle'],
    y=data['Count'],
    text=data['Count'],  # Data labels
    textposition='auto',  # Position of data labels
    marker=dict(color='rgb(71, 157, 220)'),  # Bar color
    name='Count'
))

# Create line plot with secondary y-axis
fig.add_trace(go.Scatter(
    x=data['Circle'],
    y=data['Contribution'],
    name='Contribution %',
    mode='lines',
    line=dict(color='red', dash='dash'),
    yaxis='y2'
))

# Create the second y-axis
fig.update_layout(
    yaxis2=dict(title='Contribution %', overlaying='y', side='right', showgrid=False),
)

# Set axis limits
fig.update_layout(
    xaxis=dict(range=[-0.5, len(data)-0.5]),
    yaxis=dict(range=[0, max(data['Count'])+10]),
    yaxis2=dict(range=[-5, 20])
)

# Update layout
fig.update_layout(
    title="Highest Ratings",
    xaxis=dict(title="Circle"),
    yaxis=dict(title="Count"),
)

fig.show()
```

```python
import plotly.graph_objects as go

# Get data
chains = combined_df['CIRCLE_NAME'].value_counts()
data = pd.DataFrame({'Circle': chains.index, 'Count': chains.values})

# Calculate contribution percentage
total_count = data['Count'].sum()
data['Contribution'] = (data['Count'] / total_count) * 100

# Create bar plot using Plotly
fig = go.Figure()

# Add bar trace with data labels
fig.add_trace(go.Bar(
    x=data['Circle'],
    y=data['Count'],
    text=data['Count'],  # Data labels
    textposition='auto',  # Position of data labels
    marker=dict(
        color='rgb(71, 157, 220)',  # Bar color
        line=dict(color='rgb(40, 112, 179)', width=1)  # Bar border color and width
    ),
    name='Count'
))

# Create line plot with secondary y-axis
fig.add_trace(go.Scatter(
    x=data['Circle'],
    y=data['Contribution'],
    name='Contribution %',
    mode='lines',
    line=dict(color='red', dash='dash'),
    yaxis='y2'
))

# Create the second y-axis
fig.update_layout(
    yaxis2=dict(
        title='Contribution %',
        overlaying='y',
        side='right',
        showgrid=False,
        titlefont=dict(color='red'),
        tickfont=dict(color='red')
    ),
)

# Set axis limits
fig.update_layout(
    xaxis=dict(range=[-0.5, len(data)-0.5]),
    yaxis=dict(range=[0, max(data['Count'])+10]),
    yaxis2=dict(range=[0, 20])
)

# Update layout
fig.update_layout(
    title="Highest Ratings",
    xaxis=dict(title="Circle", tickangle=270),  # Rotate x-axis labels for better readability
    yaxis=dict(title="Count"),
    plot_bgcolor='rgb(243, 243, 243)',  # Background color
    paper_bgcolor='rgb(243, 243, 243)',  # Plot area background color
    font=dict(family="Arial", size=12, color="black")  # Font style and size
)

# Add gridlines
fig.update_xaxes(showgrid=True, gridwidth=0.5, gridcolor='lightgray')
fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor='lightgray')

# Add plot annotations
fig.add_annotation(
    xref='paper', yref='paper',
    x=0.5, y=1.15,
    text='Data Source: Your Source',
    showarrow=False,
    font=dict(size=10, color='gray')
)

fig.show()
```
