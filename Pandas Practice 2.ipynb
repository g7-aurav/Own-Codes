## Data cleaning in pandas

```python
df9 = pd.read_excel(r"D:\GAURAV\My Learnings\Pandas project dataset\Project 2\Customer Call List.xlsx")
df9
```

```python
df9.drop_duplicates()
```

```python
df9 = df9.drop_duplicates()
df9
```

```python
df9 = df9.drop(columns = 'Not_Useful_Column')
df9
```

```python
df9['Last_Name'].str.strip()
```

```python
#df9['Last_Name'] = df9['Last_Name'].str.lstrip('/')
#df9['Last_Name'] = df9['Last_Name'].str.lstrip('...')
#df9['Last_Name'] = df9['Last_Name'].str.rstrip('/')
df9['Last_Name'] = df9['Last_Name'].str.strip(r'123.\\')
df9
```

```python
df9['Last_Name'] = df9['Last_Name'].str.lstrip('/')
df9
```

```python
df9
```

```python
df9['Phone_Number'] = df9['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df9
```

```python
#df9['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])   #Step 1


df9['Phone_Number'] =df9['Phone_Number'].apply(lambda x:str(x))   #Step 2

df9['Phone_Number'] = df9['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])   #Step 3

df9
```

```python
df9['Phone_Number'] = df9['Phone_Number'].str.replace('nan--','')
df9['Phone_Number'] = df9['Phone_Number'].str.replace('Na--','')
df9
```

```python
df9['Address'].str.split(',',2,expand = True) #checking the output
```

```python
df9[["Street_Address","State","Zip_Code"]] = df9['Address'].str.split(',',2,expand = True) 
df9
```

```python
df9['Paying Customer'] = df9['Paying Customer'].str.replace('Yes','Y')
df9['Paying Customer'] = df9['Paying Customer'].str.replace('No','N')
df9
```

```python
df9['Do_Not_Contact'] = df9['Do_Not_Contact'].str.replace('Yes','Y')
df9['Do_Not_Contact'] = df9['Do_Not_Contact'].str.replace('No','N')
df9
```

```python
#df9 = df9.str.replace('N/a','')
df9 = df9.replace('N/a','')
df9 = df9.replace('NaN','')
df9 = df9.fillna("")
df9
```

```python
df9[df9['Do_Not_Contact'] != 'Y']      #Easy method
```

```python
for x in df9.index:
    if df9.loc[x,"Do_Not_Contact"] == 'Y':
        df9.drop(x,inplace = True)
df9
```

```python
for x in df9.index:
    if df9.loc[x,"Phone_Number"] == '':
        df9.drop(x,inplace = True)
df9

#df9 = df9.dropna(subset = 'Phone_Number'), inplace  = True) #Alternate
```

```python
df9 = df9.reset_index(drop = True)
df9
```
