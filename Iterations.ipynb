```python
x = 15

if x > 20:
    print("X is very large")
elif x > 10:
    print("X is medium-sized")  # This runs, then the block exits
else:
    print("X is small")
```

```python
my_dict = {"apple": 3, "banana": 5, "cherry": 2}

for key, value in my_dict.items():
    print(f"We have {value} units of {key}.")
```

```python
# Loop through KEYS only (Default behavior)
my_dict = {"apple": 3, "banana": 5, "cherry": 2}
for key in my_dict:
    print(key)

# Loop through VALUES only
for value in my_dict.values():
    print(value)
```

```python
my_list = [1, 2, 3]
# Running a math operation on list items
for num in my_list:
    print(num * 10)
```

```python
# Accumulating values from a tuple
my_tuple = (4, 5, 6)
total = 0
for num in my_tuple:
    total += num
print(total)
```

```python
numbers = [5, 12, 7, 18, 3, 25]
# print "small" if < 10, "medium" if 10-20, "large" if > 20
for number in numbers:
    if number <10:
        print("small")
    elif number <= 20:
        print("Medium")
    else:
        print("large")
```
