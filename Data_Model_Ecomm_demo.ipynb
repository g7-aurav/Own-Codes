```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\GAURAV\My Learnings\Spark Tutorials Youtube\Data_Model_Creation\Practice_file.csv")
df.head(5)
```

```python
print(df.columns.tolist())
```

```python
df.columns = [

    col.strip()
       .lower()
       .replace(' ', '_')
       .replace('.', '')
       .replace("'", '')
    
    for col in df.columns

]

print(df.columns.tolist())
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'])
```

```python
dim_customer = df[[

    'cust_id',
    'name_prefix',
    'first_name',
    'middle_initial',
    'last_name',
    'gender',
    'e_mail',
    'fathers_name',
    'date_of_birth',
    'ssn',
    'phone_no'

]].drop_duplicates().reset_index(drop=True)
```

```python
dim_customer['customer_key'] = range( 1, len(dim_customer)+1 ) 
dim_customer.head()
```

```python
dim_location = df[[

    'region',
    'state',
    'city',
    'zip'

]].drop_duplicates().reset_index(drop=True)
```

```python
dim_location['location_key'] = range( 1, len(dim_location)+1 ) 
dim_location.head()
```

```python
dim_product = df[[ 'product_name' ]].drop_duplicates().reset_index(drop=True)
```

```python
dim_product['product_key'] = range( 1, len(dim_product)+1 ) 
dim_product.head()
```

```python
all_dates = pd.concat([ df['order_date'], df['ship_date'] ]).drop_duplicates()
```

```python
dim_date = pd.DataFrame({ 'full_date': all_dates })
```

```python
dim_date['date_key'] = dim_date[ 'full_date' ].dt.strftime('%Y%m%d').astype(int) 
dim_date['year'] = dim_date[ 'full_date' ].dt.year 
dim_date['month'] = dim_date[ 'full_date' ].dt.month 
dim_date['day'] = dim_date[ 'full_date' ].dt.day 
dim_date['weekday'] = dim_date[ 'full_date' ].dt.day_name() 
dim_date['quarter'] = dim_date[ 'full_date' ].dt.quarter
```

```python
dim_date.head()
```

```python
fact_sales = df.copy()
```

```python
fact_sales = fact_sales.merge( dim_customer[ ['cust_id', 'customer_key'] ], on='cust_id', how='left' )
```

```python
fact_sales = fact_sales.merge( dim_product[ ['product_name', 'product_key'] ], on='product_name', how='left' )
```

```python
fact_sales = fact_sales.merge( dim_location[ [ 'region', 'state', 'city', 'zip', 'location_key' ] ], on=[ 'region', 'state', 'city', 'zip' ], how='left' )
```

```python
fact_sales['order_date_key'] = fact_sales[ 'order_date' ].dt.strftime('%Y%m%d').astype(int)
```

```python
fact_sales['ship_date_key'] = fact_sales[ 'ship_date' ].dt.strftime('%Y%m%d').astype(int)
```

```python
fact_sales['total_sales'] = ( fact_sales['units_sold'] * fact_sales['unit_price'] )
```

```python
fact_sales = fact_sales[[ 'order_id', 'customer_key', 'product_key', 'location_key', 'order_date_key', 'ship_date_key', 'units_sold', 'unit_price', 'total_sales' ]]
```

```python
fact_sales['sales_key'] = range( 1, len(fact_sales)+1 )
```

```python
fact_sales = fact_sales[[ 'sales_key', 'order_id', 'customer_key', 'product_key', 'location_key', 'order_date_key', 'ship_date_key', 'units_sold', 'unit_price', 'total_sales' ]] 
fact_sales.head()
```

```python
print('\nDIM CUSTOMER')
print(dim_customer.head())

print('\nDIM LOCATION')
print(dim_location.head())

print('\nDIM PRODUCT')
print(dim_product.head())

print('\nDIM DATE')
print(dim_date.head())

print('\nFACT SALES')
print(fact_sales.head())
```
