```python
# Write a function that takes a name and returns "Hello, <name>!"
def call_me(name):
    output = "hello" + " " + name 
    return output
print(call_me("gaurav"))
```

```python
# Write a function that takes price and quantity, returns total cost
def total_cost(price,quantity):
    final_amount = price*quantity
    return final_amount

print(total_cost(150,11)) 
```

```python
# Write a function that takes a number and returns "even" or "odd"
def eve_or_odd(num):
    result = "even" if num % 2 == 0 else "odd"
    return result
print(eve_or_odd(10))
```

```python
# Write a function that takes a list of numbers and returns the average
# Handle the case where the list is empty
def cal_avg(num):
    result =  0 if not num else sum(num)/len(num)
    return result
print(cal_avg([2,3,4,56]))  
```

```python
# Write a function that divides two numbers
# Use try/except inside it to handle division by zero
try:
    def div(a,b):
        output = a/b
        return output
    print(div(5,0))
except ZeroDivisionError:
    print("Division by zero")
```

```python
# Write a function that divides two numbers
# Use try/except inside it to handle division by zero
try:
    def div(a,b):
        output = a/b
        return output
    print(div(a,2))
except Exception as e:
    print(e)
```

```python
# Write a function that reads a number from a string input
# Return the number if valid, return 0 if invalid
# Use try/except inside
def parse_number(string_input):
    try:
        return float(string_input)
    except ValueError:
        return 0
# Put the inputs inside a list as strings
inputs = ["a", "2", "e", "4"]

for item in inputs:
    print(parse_number(item))
```
