```python
import pandas as pd

input_file_path = r'C:\Users\QD2530\Desktop\New folder\01032024.csv'
output_file_path = r'C:\Users\QD2530\Desktop\New folder\01032024 Sampler1.csv'

df = pd.read_csv(input_file_path)

filter1 = df['Buckets'] == '>=1 M 30 S'
filtered_df = df[filter1]

# Randomly shuffle the DataFrame
shuffled_df = filtered_df.sample(frac=1)

# Create a new DataFrame to hold duplicated rows
duplicated_rows = []

# Loop through unique agents
for agent in shuffled_df['Agent Name'].unique():
    agent_df = shuffled_df[shuffled_df['Agent Name'] == agent]
    
    # For each date of the agent, duplicate two records (if available)
    for date in agent_df['Recording date'].unique():
        date_df = agent_df[agent_df['Recording date'] == date]
        duplicated_rows.extend(date_df.head(5).values.tolist())  # Selecting two records for each date

# Create a new DataFrame from the duplicated rows
duplicated_df = pd.DataFrame(duplicated_rows, columns=shuffled_df.columns)

# Save the DataFrame with duplicated rows to a CSV file
duplicated_df.to_csv(output_file_path, index=False)
```
