```python
product_name = []
prices = []
description = []
reviews = []

for i in range(2,12) :
    url = "https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page=" + str(i)
    r = requests.get(url)

    soup = BeautifulSoup(r.text, "lxml")
    box = soup.find("div", class_="_1YokD2")

    names = box.find_all("div", class_="_4rR01T")
    for i in names:
        name = i.text
        product_name.append(name)

    price = box.find_all("div", class_="_30jeq3 _1_WHN1")
    for i in price:
        price = i.text
        prices.append(price)

    desc = box.find_all("div", class_="fMghEO")
    for i in desc:
        desc = i.text
        description.append(desc)

    review = box.find_all("div", class_="_3LWZlK")
    for i in review:
        review = i.text
        reviews.append(review)

    #lengths of lists and handle missing values
    max_length = max(len(product_name), len(prices), len(description), len(reviews))
    product_name += ['0'] * (max_length - len(product_name))
    prices += ['0'] * (max_length - len(prices))
    description += ['0'] * (max_length - len(description))
    reviews += ['0'] * (max_length - len(reviews))

df1 = pd.DataFrame({"Product Name": product_name, "Prices": prices, "Description": description, "Review": reviews})
df1.head()
```
