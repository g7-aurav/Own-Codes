```python
# Customize the figure size
plt.figure(figsize=(15, 5))

# Create a countplot
ax = sns.countplot(x='Attrition', hue='JobSatisfaction', data=HRA, palette='flag')

# Add data labels (annotations)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', fontsize=15, color='black', xytext=(0, 10), textcoords='offset points')

# Add labels and title
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.title('Countplot of Attrition with JobSatisfaction Hue')

# Show the plot
plt.show()
```
