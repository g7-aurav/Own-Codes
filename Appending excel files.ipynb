```python
import pandas as pd

# File paths for the two Excel files
file_path1 = r'C:\GAURAV\ExcelR\Python append practice\Dimcustomer(1).xlsx'
file_path2 = r'C:\GAURAV\ExcelR\Python append practice\Dimcustomer(2).xlsx'

# Load the Excel files into DataFrames
data1 = pd.read_excel(file_path1, sheet_name='Dimcustomer')
data2 = pd.read_excel(file_path2, sheet_name='Dimcustomer')

# Filter data for 'Clerical' in the 'EnglishOccupation' column
data1_clerical = data1[data1['EnglishOccupation'] == 'Clerical']
data2_clerical = data2[data2['EnglishOccupation'] == 'Clerical']

# Concatenate the filtered DataFrames
combined_data = pd.concat([data1_clerical, data2_clerical], axis=0)

# Save the combined DataFrame to a new Excel file
combined_file_path = r'C:\GAURAV\ExcelR\Python append practice\combine1.xlsx'
# Assuming the intention to write to a new sheet or override Sheet1
combined_data.to_excel(combined_file_path, sheet_name='Sheet1', index=False)
```

```python
import pandas as pd

# File paths for the two Excel files
file_path1 = r'C:\GAURAV\ExcelR\Python append practice\Dimcustomer(1).xlsx'
file_path2 = r'C:\GAURAV\ExcelR\Python append practice\Dimcustomer(2).xlsx'

# Load the Excel files into DataFrames
data1 = pd.read_excel(file_path1, sheet_name='Dimcustomer')
data2 = pd.read_excel(file_path2, sheet_name='Dimcustomer')

# Filter data for 'Clerical' in the 'EnglishOccupation' column and 'Bachelors' in the 'EnglishEducation' column
data1_filtered = data1[(data1['EnglishOccupation'] == 'Clerical') & (data1['EnglishEducation'] == 'Bachelors')]
data2_filtered = data2[(data2['EnglishOccupation'] == 'Clerical') & (data2['EnglishEducation'] == 'Bachelors')]

# Concatenate the filtered DataFrames
combined_data = pd.concat([data1_filtered, data2_filtered], axis=0)

# Save the combined DataFrame to a new Excel file
combined_file_path = r'C:\GAURAV\ExcelR\Python append practice\combine.xlsx'
# Write to Sheet1, overwriting it if it exists
combined_data.to_excel(combined_file_path, sheet_name='Sheet1', index=False)
```

```python
import pandas as pd
import os

# Specify the folder containing the Excel files
folder_path = r'C:\GAURAV\ExcelR\Python append practice'

# List all Excel files in the specified folder
excel_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]

# Initialize an empty list to store DataFrames
dataframes = []

# Loop through the list of Excel files
for file in excel_files:
    # Create the full file path
    file_path = os.path.join(folder_path, file)
    # Load the Excel file into a DataFrame
    df = pd.read_excel(file_path, sheet_name='Sheet1')
    # Append the DataFrame to the list
    dataframes.append(df)

# Concatenate all DataFrames from the list
combined_data = pd.concat(dataframes, axis=0)

# Save the combined DataFrame to a new Excel file
combined_file_path = os.path.join(folder_path, 'combine.xlsx')
combined_data.to_excel(combined_file_path, sheet_name='Sheet1', index=False)
```
