```python
import os
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("MyApp") \
    .master("local[*]") \
    .getOrCreate()
```

```python
print("Master:", spark.sparkContext.master)
print("Default Parallelism:", spark.sparkContext.defaultParallelism)
print("CPU Cores Available:", os.cpu_count())
```

```python
rdd = spark.sparkContext.parallelize(range(1000))
print(rdd.getNumPartitions())
```

```python
for k, v in spark.sparkContext.getConf().getAll():
    print(f"{k} = {v}")
```

```python
import os

print("Spark Version:", spark.version)
print("Master:", spark.sparkContext.master)
print("App Name:", spark.sparkContext.appName)
print("Default Parallelism:", spark.sparkContext.defaultParallelism)
print("CPU Threads:", os.cpu_count())
print("Driver Memory:", spark.conf.get("spark.driver.memory", "Not Set"))
print("Executor Memory:", spark.conf.get("spark.executor.memory", "Not Set"))
print("Shuffle Partitions:", spark.conf.get("spark.sql.shuffle.partitions"))
print("AQE Enabled:", spark.conf.get("spark.sql.adaptive.enabled"))
print("Spark UI:", spark.sparkContext.uiWebUrl)
```

```python
df = spark.range(1000000)

print("Partitions:", df.rdd.getNumPartitions())
```

```python
df2 = spark.range(1000000).repartition(32)

print(df2.rdd.getNumPartitions())
```
