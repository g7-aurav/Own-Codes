```python
def assign_cost_centre(row):
    if row['Service Code'] == 'WRP' and row['Provider  Province/State'] == 'California':
        return 'C4199'
    elif row['Service Code'] == 'SA' and row['Provider  Province/State'] == 'California':
        return 'C4199'
    elif row['Service Code'] == 'ST2' and row['Provider  Province/State'] == 'California':
        return 'C4199'
    elif row['Service Code'] == 'ST' and row['Provider  Province/State'] == 'California':
        return 'C4199'
    elif row['Service Code'] == 'AR' and row['Provider  Province/State'] == 'California':
        return 'C4199'
    elif row['Service Code'] == 'WRP':
        return 'C4103'
    elif row['Service Code'] == 'SA':
        return 'C4103'
    elif row['Service Code'] == 'ST2':
        return 'C4156'
    elif row['Service Code'] == 'ST':
        return 'C4156'
    elif row['Service Code'] == 'AR':
        return 'C4156'
    elif row['Service Code'] == 'Knox Keene':
        return 'C4199'
    elif row['Service Code'] == 'SM':
        return 'C4305'
    elif row['Service Code'] == 'WSH':
        return 'C4305'
    elif row['Service Code'] == 'HLC':
        return 'C4507'
    elif row['Service Code'] == 'SC':
        return 'C4507'
    elif row['Service Code'] == 'SL':
        return 'C4507'
    elif row['Service Code'] == 'LC':
        return 'C4507'
    elif row['Service Code'] == 'DBR':
        return 'C4120'
    else :
        return 'C4120'

New_data['Cost Centre'] = New_data.apply(assign_cost_centre,axis = 1)
```

```python
import numpy as np

# Define your conditions in order of priority
conditions = [
    # California-specific
    ((New_data['Service Code'].isin(['WRP', 'SA', 'ST2', 'ST', 'AR'])) &
     (New_data['Provider  Province/State'] == 'California')),

    # Non-California but service codes
    (New_data['Service Code'].isin(['WRP', 'SA'])),
    (New_data['Service Code'].isin(['ST2', 'ST', 'AR'])),
    (New_data['Service Code'] == 'Knox Keene'),
    (New_data['Service Code'].isin(['SM', 'WSH'])),
    (New_data['Service Code'].isin(['HLC', 'SC', 'SL', 'LC'])),
    (New_data['Service Code'] == 'DBR')
]

# Corresponding outputs
choices = [
    'C4199',  # California specific WRP/SA/ST2/ST/AR
    'C4103',  # WRP or SA
    'C4156',  # ST2, ST, AR
    'C4199',  # Knox Keene
    'C4305',  # SM, WSH
    'C4507',  # HLC, SC, SL, LC
    'C4120'   # DBR
]

# Apply np.select with a default
New_data['Cost Centre'] = np.select(conditions, choices, default='C4120')
```
